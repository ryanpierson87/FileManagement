```python
import requests
from bs4 import BeautifulSoup

url = requests.get('http://www.kacl780.net/frasier/transcripts/season_1/episode_1/the_good_son.html')
soup = BeautifulSoup(url.content, 'html.parser')
def character_retrieve(chicken_noodle):
    direction = []

    for i in chicken_noodle.find_all('pre'):
        for j in i.find_all('b'):
            if len(str(j)) < 17 and 'Credits' not in str(j):
                #print(len(str(j)))
                direction.append(str(j))

    return direction
```
